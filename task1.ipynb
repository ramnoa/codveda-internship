```python
import pandas as pd
```

#**1.LOADING DATA**


```python

df = pd.read_excel('/content/fashion_boutique_dataset.xlsx')
df.head()
```

#**2.DATA CLEANING**

```python
df.info()
```

# **HANDLING MISSING VALUES**

```python
# Fill missing size with "Unknown"

df['size'] = df['size'].fillna("Unknown")
df.head()
```

```python
# Fill missing customer_rating with mean rating of the same category
df['customer_rating'] = df.groupby('category')['customer_rating'].transform(
    lambda x: x.fillna(x.mean())
)
df.head()
```

```python
# Fill missing return_reason with "Not Returned" where is_returned is False

df.loc[(df['return_reason'].isnull()) & (df['is_returned'] == False), 'return_reason'] = "Not Returned"
df.head()
```

```python
df.info()
```

# **REMOVING DUPLICATE VALUES**

```python

df = df.drop_duplicates()
df.info()
```

# **STANDARDIZING CATEGORICAL TEXT COLUMNS (title case)**

```python

categorical_cols = ['category', 'brand', 'season', 'size', 'color', 'return_reason']
for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
df.head()
```

# ** ENSURING PURCHASE_DATE IS DATETIME**

```python
df['purchase_date'] = pd.to_datetime(df['purchase_date'])
```

```python
df.head()
```

#**3.SAVING CLEANED DATA**

```python
df.to_excel("fashion_boutique_dataset_cleaned.xlsx", index=False)
```

```python
print("✅ Dataset cleaned and saved as 'fashion_boutique_dataset_cleaned.xlsx'")
```
